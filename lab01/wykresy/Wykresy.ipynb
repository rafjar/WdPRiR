```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
methods = []

serial = pd.read_csv('Mandelbrot_serial.csv', names=('ImageSize', 'Time'))
serial['Method'] = 'Serial'
methods.append(serial)

parallel = pd.read_csv('Mandelbrot_parallel.csv', names=('ImageSize', 'Time'))
parallel['Method'] = 'Parallel'
methods.append(parallel)

pool4 = pd.read_csv('Mandelbrot_pool_4.csv', names=('ImageSize', 'Time'))
pool4['Method'] = 'Pool 4'
methods.append(pool4)

pool8 = pd.read_csv('Mandelbrot_pool_8.csv', names=('ImageSize', 'Time'))
pool8['Method'] = 'Pool 8'
methods.append(pool8)

pool16 = pd.read_csv('Mandelbrot_pool_16.csv', names=('ImageSize', 'Time'))
pool16['Method'] = 'Pool 16'
methods.append(pool16)

pool32 = pd.read_csv('Mandelbrot_pool_32.csv', names=('ImageSize', 'Time'))
pool32['Method'] = 'Pool 32'
methods.append(pool32)

pool64 = pd.read_csv('Mandelbrot_pool_64.csv', names=('ImageSize', 'Time'))
pool64['Method'] = 'Pool 64'
methods.append(pool64)

df = pd.concat(methods)
df
```

```python
plt.figure(figsize=(12, 7))
sns.lineplot(x='ImageSize', y='Time', hue='Method', data=df)
plt.grid()
```

```python
plt.figure(figsize=(12, 7))
sns.lineplot(x='ImageSize', y='Time', hue='Method', data=df)
plt.xscale('log')
plt.yscale('log')
plt.grid()
```
